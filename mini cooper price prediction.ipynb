```python
mini = df1["info:"].str.split("\n", expand=True)
mini.insert(0, "car model", df1["title: "])
mini.drop(10, inplace=True, axis=1)
# define the new column names

new_columns = ['car model', 'price', 'mileage', 'transmission', 'year', 'fuel', 'power', 'loan', 'loan_request', 'insurance', 'insurance_premium']

#rename the columns
mini.columns = new_columns


# select the rows that match the mask
mask = ~mini['mileage'].str.contains('km')
selected_rows = mini[mask]
selected_rows.loc[:, 'mileage'] = ''

# print the selected rows
selected_rows.head(100)
# delete all rows with column 'mile' don't have value km
indexAge = selected_rows.index
mini.drop(indexAge , inplace=True)

#clean the wrong data scrap from auto24
a = selected_rows.iloc[:,[2,3,4,5,6,7,8,9,10]] 
b = selected_rows.iloc[:,[0,1]] 
a = a.shift(-1, axis=1)
c= pd.concat([a, b], axis=1, join='inner')
c= c[['car model', 'price', 'mileage', 'transmission', 'year', 'fuel', 'power', 'loan', 'loan_request', 'insurance', 'insurance_premium']]
c.loc[:, 'insurance_premium'] = 'None'
d=[mini,c]
d= pd.concat(d)
d.head(500)
```

## Merge two dataframe into one table and clean the data

```python
# rename the "miles" column to "mileage" in df1
new2 = new2.rename(columns={'miles': 'mileage','fuels': 'fuel'})

# merge the two dataframes based on the columns that match
merged_df = pd.merge(new2, d, on=['car model', 'price', 'mileage', 'year', 'fuel', 'transmission'], how='outer')

# view the merged dataframe
merged_df.head(10000)

# clean the "car model" column
merged_df["car model"] = merged_df["car model"].str.replace("\n", " ")  # remove the \n

# clean the "price" column
merged_df["price"] = merged_df["price"].str.replace("€", "")  # remove the euro sign
merged_df["price"] = merged_df["price"].str.replace(",", "")  # remove the comma separator
merged_df["price"] = merged_df["price"].str.replace("-", "")  # remove the dash
merged_df["price"] = merged_df["price"].str.replace(".", "")  # remove the dot

# clean the "mileage" column
merged_df["mileage"] = merged_df["mileage"].str.replace("km", "")  # remove the euro sign
merged_df["mileage"] = merged_df["mileage"].str.replace(".", "")  # remove the dot

# clean the "year" column
merged_df["year"] = merged_df["year"].str.extract(r"(\d{4})")

# clean the "fuel" column
merged_df["fuel"] = merged_df["fuel"].str.replace("Benzine", "Petrol") # translate Duth into English


# clean the "transmission" column
merged_df["transmission"] = merged_df["transmission"].str.replace("Cabriolet",  "")
merged_df["transmission"] = merged_df["transmission"].str.replace("Automatisch",  "")
merged_df["transmission"] = merged_df["transmission"].str.replace("Hatchback",  "")
merged_df["transmission"] = merged_df["transmission"].str.replace("Overige carrosserieën",  "")
merged_df["transmission"] = merged_df["transmission"].str.replace("Transmissie",  "")
merged_df["transmission"] = merged_df["transmission"].str.replace("Half/Semi-automaat",  "")
merged_df["transmission"] = merged_df["transmission"].str.replace("-None", "")
merged_df["transmission"] = merged_df["transmission"].str.replace('- ', "")
# convert the column to numeric values
merged_df["price"] = pd.to_numeric(merged_df["price"], errors="coerce")
merged_df["year"] = pd.to_numeric(merged_df["year"], errors="coerce")
merged_df["mileage"] = pd.to_numeric(merged_df["mileage"], errors="coerce")
# display the cleaned data
merged_df.head(10000)
```

```python
print(merged_df['warranty'].unique())
print(merged_df['fuel'].unique())
print(merged_df['transmission'].unique())
```

```python
#clean the wrong data in the column fuel
class_23 = merged_df[merged_df["fuel"].isin(["Handgeschakeld","Automaat",'Hatchback','Overige carrosserieën'])]
class_23.head(100)
indexfuel = class_23.index
merged_df.drop(indexfuel , inplace=True)
# clean the outrange price
low_price = merged_df[merged_df["price"]<100]
index_lowprice = low_price.index
merged_df.drop(index_lowprice , inplace=True)
# clean the outrange year
low_year = merged_df[merged_df["year"]<1800]
index_lowyear = low_year.index
merged_df.drop(index_lowyear , inplace=True)

# categorize the car models based on their name
merged_df.loc[merged_df["car model"].str.contains("mini cooper", case=False), "category"] = "Mini Cooper"
merged_df.loc[merged_df["car model"].str.contains("cooper s", case=False), "category"] = "Cooper S"
merged_df.loc[merged_df["category"] == "", "category"] = "Others"
#clean power columns
merged_df['power'] = merged_df['power'].str.replace('kw|pk|[(]|[)]', '', regex=True)
merged_df['power'] = merged_df['power'].apply(lambda x: re.findall('\d+', str(x))[0] if len(re.findall('\d+', str(x))) > 0 else None)
merged_df["power"] = pd.to_numeric(merged_df["power"], errors="coerce")
merged_df['power'] = merged_df['power'].replace(np.nan, 0)
#delete useless columns
merged_df.drop('loan', inplace=True, axis=1)
merged_df.drop('loan_request', inplace=True, axis=1)
merged_df.drop('insurance', inplace=True, axis=1)
merged_df.drop('insurance_premium', inplace=True, axis=1)
merged_df.drop('car model', inplace=True, axis=1)
# clean the duplicates data
result_df = merged_df.drop_duplicates(keep='last')
result_df.describe()
result_df.head(100)
```

```python
result_df_expanded = pd.get_dummies(result_df)
result_df_expanded = result_df_expanded.dropna()
result_df_expanded.head(10000)
```

```python
#Applying the standard scalar option to standardize all the variables in the dataset.
std = StandardScaler()
result_df_expanded_std = std.fit_transform(result_df_expanded)
result_df_expanded_std = pd.DataFrame(result_df_expanded_std, columns = result_df_expanded.columns)
print(result_df_expanded_std.shape)
result_df_expanded_std.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(result_df_expanded_std.drop(columns = ['price']), result_df_expanded_std[['price']])
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
column_names = result_df_expanded.drop(columns = ['price']).columns

no_of_features = []
r_squared_train = []
r_squared_test = []

for k in range(3, 20, 2):
    selector = SelectKBest(f_regression, k = k)
    X_train_transformed = selector.fit_transform(X_train, y_train)
    X_test_transformed = selector.transform(X_test)
    regressor = LinearRegression()
    regressor.fit(X_train_transformed, y_train)
    no_of_features.append(k)
    r_squared_train.append(regressor.score(X_train_transformed, y_train))
    r_squared_test.append(regressor.score(X_test_transformed, y_test))
    
sns.lineplot(x = no_of_features, y = r_squared_train, legend = 'full')
sns.lineplot(x = no_of_features, y = r_squared_test, legend = 'full')
```

```python
selector = SelectKBest(f_regression, k = 20)
X_train_transformed = selector.fit_transform(X_train, y_train)
X_test_transformed = selector.transform(X_test)
column_names[selector.get_support()]
```

```python
def regression_model(model):
    """
    Will fit the regression model passed and will return the regressor object and the score
    """
    regressor = model
    regressor.fit(X_train_transformed, y_train)
    score = regressor.score(X_test_transformed, y_test)
    return regressor, score
```

```python
model_performance = pd.DataFrame(columns = ["Features", "Model", "Score"])

models_to_evaluate = [LinearRegression(), Ridge(), Lasso(), SVR(), RandomForestRegressor(), MLPRegressor()]

for model in models_to_evaluate:
    regressor, score = regression_model(model)
    model_performance = model_performance.append({"Features": "Linear","Model": model, "Score": score}, ignore_index=True)

model_performance
```
